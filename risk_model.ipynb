```python
# 1. Sort encounters by patient and date
encounters_sorted = encounters_df.sort_values(['PATIENT', 'START'])

# 2. Convert START to datetime if not already
encounters_sorted['START'] = pd.to_datetime(encounters_sorted['START'])

# 3. Calculate efficient costs without creating multi-index
def calculate_efficient_costs(group):
    
    # Calculate total future costs (not including current encounter)
    group['TOTAL_FUTURE_COST'] = group['TOTAL_CLAIM_COST'].shift(-1)[::-1].cumsum()[::-1]
    group['TOTAL_FUTURE_COVERAGE'] = group['PAYER_COVERAGE'].shift(-1)[::-1].cumsum()[::-1]
    
    return group

# 4. Apply calculations to each patient group and reset index
future_costs_df = encounters_sorted.groupby('PATIENT', group_keys=False).apply(calculate_efficient_costs)

future_costs_df['FUTURE_OUT_OF_POCKET'] = future_costs_df['TOTAL_FUTURE_COST'] - future_costs_df['TOTAL_FUTURE_COVERAGE']

# 5. Sort and select relevant columns
encounters_info = future_costs_df[[
    'PATIENT', 
    'Id', 
    'START', 
    'TOTAL_CLAIM_COST',
    'TOTAL_FUTURE_COST',
    'PAYER_COVERAGE',
    'TOTAL_FUTURE_COVERAGE',
    'FUTURE_OUT_OF_POCKET',
]].sort_values(['PATIENT', 'START'])

# Display results
display(encounters_info)
```

```python
# augment encounter_info with the sum of every Total_Claim_Cost and Payer_Coverage (needed to calculate out of pocket) where the start date > current start and patient is the same

pivot_table = observations_df.pivot_table(
    index='ENCOUNTER',
    columns='DESCRIPTION',
    values='VALUE',
    aggfunc='first'  # Takes the first value if there are duplicates
).reset_index()

flattened_observations_df = pd.merge(
    encounters_info,
    pivot_table,
    left_on='Id',
    right_on='ENCOUNTER',
    how='right'
)

flattened_observations_df.drop(columns='Id', inplace=True)
flattened_observations_df.rename(columns={'START': 'DATE'}, inplace=True)

display(flattened_observations_df)
```

```python
conditions_df['PRESENT'] = 1

flattened_conditions_df = conditions_df.pivot_table(
    index='ENCOUNTER',
    columns='DESCRIPTION',
    values='PRESENT',
    aggfunc='first', # Takes the first value if there are duplicates
    fill_value=0,
).reset_index()

display(flattened_conditions_df)
```

```python
combined_df = pd.merge(
    flattened_observations_df,
    flattened_conditions_df,
    on='ENCOUNTER',
    how='inner'
)

display(combined_df)
```

```python
# Group by PATIENT and count occurrences, then filter to keep only patients with < 5 appearances
patient_counts = combined_df.groupby('PATIENT').size()
patients_to_keep = patient_counts[patient_counts > 4].index

# Filter the original dataframe to keep only rows where PATIENT is in the filtered list
final_df = combined_df[combined_df['PATIENT'].isin(patients_to_keep)]

final_df.loc[:, "DATE"] = pd.to_datetime(final_df["DATE"], errors='coerce').dt.tz_localize(None)
comparison_date = pd.to_datetime('2000-01-01')
final_df = final_df[final_df["DATE"] >= comparison_date]

display(final_df)
```

```python
filtered_patients_df = patients_df[['Id', 'BIRTHDATE', 'RACE', 'ETHNICITY', 'GENDER']]

final_df = pd.merge(
    filtered_patients_df,
    final_df,
    left_on='Id',
    right_on='PATIENT',
    how='inner',
)

final_df.drop(columns='Id', inplace=True)

display(final_df)
```

```python
final_df.loc[:, "DATE"] = pd.to_datetime(final_df["DATE"], errors='coerce')

final_df.loc[:, "BIRTHDATE"] = pd.to_datetime(final_df["BIRTHDATE"], errors='coerce')

final_df['AGE'] = final_df.apply(lambda row: row['DATE'].year - row['BIRTHDATE'].year - ((row['DATE'].month, row['DATE'].day) < (row['BIRTHDATE'].month, row['BIRTHDATE'].day)), axis=1)
display(final_df)
```
